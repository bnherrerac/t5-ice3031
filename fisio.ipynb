```python
fig, ax = plt.subplots(figsize=(6,5))
# ax.plot(1, np.mean(post_pam), color='red', label='Después (0 min)')
# ax.plot(2, np.mean(treinta_pam), color='orange', label='Después (30 min)')
ax.errorbar(-5, np.mean(reposo_pam), yerr=np.std(reposo_pam), color='dimgrey')
ax.errorbar(0, np.mean(post_pam), yerr=np.std(post_pam), color='dimgrey')
ax.errorbar(30, np.mean(treinta_pam), yerr=np.std(treinta_pam), color='dimgrey')
ax.plot([-5,0,30], [np.mean(reposo_pam), np.mean(post_pam), np.mean(treinta_pam)], '--', marker='o', color='blue')
fig.suptitle(r"Promedio de presión arterial media (PAM) en 3 sujetos"+"\n"+ "en distintas etapas de esfuerzo físico.", fontsize=15)

# ax.set_xticks([0,1,2],[-5,0,30], fontsize=15)
# ax.set_yticklabels(fontsize=15)
ax.set_ylabel(r"PAM = $P_{d}+\frac{1}{3}(P_{s}-P_{d})$ [mmHg]", fontsize=13)
ax.set_xlabel('Tiempo desde el final del ejercicio [min]', fontsize=13)
ax.tick_params(labelsize=14)
ax.grid(True)
plt.tight_layout()
plt.show()
print(np.mean(reposo_pam), np.mean(post_pam), np.mean(treinta_pam))
print(np.std(reposo_pam),np.std(post_pam),np.std(treinta_pam))
print((np.mean(post_pam)-np.mean(reposo_pam))/5)
```
